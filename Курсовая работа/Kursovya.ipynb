```python
#–ø–µ—Ä–≤–∞—è –∑–∞–¥–∞—á–∞ –ø—Ä–æ —Ä–∞–≤–Ω—ã–π —Å—á–µ—Ç –≤—Å–µ—Ö –∫–æ–º–∞–Ω–¥ —Å –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–º–∏ –Ω–∏—á—å–∏–º–∏
N = int(input())
for i in range(N):
    n = int(input())
    if n%2 == 0:
        ret = [1]*(n//2-1)+[0]+[-1]*(n//2-1)
    else:
        ret = [1]*(n//2)+[-1]*(n//2)
 
    for i in range(len(ret),0,-1):
        print(*ret[:i],end=' ')
    print()  
```

```python
#—Ç—Ä–µ—Ç—å—è –∑–∞–¥–∞—á–∞ –ª–æ–≥–∏—á–µ—Å–∫–∞—è —Å—É–º–º–∞ –Ω–∞–±–æ—Ä–æ–≤ –ø–∞—Ä
def solve(n, k):
    result = []
    for i in range(n // 2):
        result.append([i, n - 1 - i])
    if k == n - 1:
        result[0][0] = n - 2
        result[1][1] = n - 3
        result[2][1] = 0
        return result
    replace_index = min(k, n - 1 - k)
    if replace_index == k:
        result[replace_index][0] = 0
    else:
        result[replace_index][1] = 0
    result[0][0] = k
 
    return result
 
t = int(input())
 
for _ in range(t):
    n, k = [int(x) for x in input().split()]
    if k == n - 1 and n == 4:
        print(-1)
    else:
        result = solve(n, k)
        for i in result:
            print(i[0], i[1])
        
```

```python
#—á–µ—Ç–≤–µ—Ä—Ç–∞—è –∑–∞–¥–∞—á–∞ –í–∞—Å—è –∏ –ü–µ—Ç—è —É–≥–∞–¥—ã–≤–∞—é—Ç —á–∏—Å–ª–∞
n = int(input())
 
def is_prime(x):
    if x == 2:
        return True
    if x == 1 or x % 2 == 0:
        return False
    i = 3
    while i * i <= x:
        if x % i == 0:
            return False
        i += 1
    return True
 
ans = []
for i in range(2, n + 1):
    if is_prime(i):
        j = i
        while j <= n:
            ans.append(j)
            j *= i
 
 print(len(ans))
print(' '.join(map(str, ans)))
```

```python
#—à–µ—Å—Ç–∞—è –∑–∞–¥–∞—á–∞
mod=1000000007
fact=[1] 
def inv(a):
    return pow(a,mod-2,mod)
def C(n,k):
    if(k>n):
        return 0
    return fact[n] * inv((fact[k]*fact[n-k])%mod) %mod
def fac(n,p):
    for i in range(p,n+1):
        fact.append((fact[i-1]*i)%mod)
        p=p+1
    return p
def initialize(n,p):
    if n>p:
        return fac(n,p)
    else:
        return p
t=int(input())
p=1
for _ in range(t):
    n,m,k=map(int,input().split())
    a=list(map(int,input().split()))
    p=initialize(n,p)
    a.sort()
    l=0
    s=0
    #print(p)
    for j in range(n):
        while a[j]-a[l]>k and l<=j:
            l=l+1
        s=(s+C(j-l,m-1))%mod
    print(s)
#print(fact)
```
